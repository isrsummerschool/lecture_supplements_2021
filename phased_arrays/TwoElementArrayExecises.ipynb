```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This is the simple two-element pattern from the phasors lecture
theta=np.linspace(0,2*np.pi,1000)

I1=1.0
I2=np.exp(1j*np.pi/2.0)
kd=np.pi/2
Fsq=np.abs(I1+I2*np.exp(1j*kd*np.cos(theta)))**2.0

#polar plot
plt.rcParams['figure.figsize']=10,10
plt.rcParams['font.size']=18

fig,ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta,Fsq,linewidth=3.0)
ax.set_rmin(0.0)
plt.show()
```
